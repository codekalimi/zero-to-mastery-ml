## Matplotlib example workflow

```python
# 0. import matplotlib and get it ready for plotting in Jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare data
x = [1,2,3,4]
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize=(10,10)) # (width, height)

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & show (you save the whole figure)
fig.savefig("../images/sample-plot-created.png")
```

```python
x_data = [1,2,3,4,5]
y_data = [4,2,3,4,1]
fig, ax = plt.subplots(figsize=(10,10))
ax.plot(x_data, y_data)
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")
fig.savefig("../images/sample-plot-created.png")
```

## Making figures with NumPy arrays

```python
# create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
x
```

```python
# Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2);
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis");
```

```python
fig, ax = plt.subplots()
ax.plot(x, x ** 2);
ax.set(title="Example",
       xlabel="x data",
       ylabel="y data");
```

Keep in Mind: If we are in a different shell we need to again call fig, ax = plt.subplots() to create a new figure and axes to plot on.

```python
# Use same data to make a scatter
fig, ax = plt.subplots()
ax.scatter(x, x ** 2);
ax.set(title="Example",
       xlabel="x data",
       ylabel="y data");
```

```python
# make a plot from dictionary
nut_butter_prices = {"Almond butter": 10,
                     "Peanut butter": 8,
                     "Cashew butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax.set(title="Nut Butter Store",
       xlabel="Butters",
       ylabel="Price");
```

```python
# lets look horizontally
fig, ax = plt.subplots()
ax.barh(list(nut_butter_prices.keys()), list(nut_butter_prices.values()));
ax.set(title="Nut Butter Store",
       xlabel="Prices",
       ylabel="Butters");
```

```python
# make some data for histograms and plot it
fig, ax = plt.subplots()
ax.hist(nut_butter_prices.values());
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.hist(x);
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.plot(x);
```

```python
# Subplots option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                             ncols=2,
                                             figsize=(10, 5))
ax1.plot(x, x/2);
ax2.scatter(x, x/2);
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(x);
# set the title for the figure, x and y axes
fig.suptitle("Random Plotting", fontsize=16, fontweight="bold");
# set x axis title, y axis title
ax1.set(title="Line Plot",
        xlabel="x-axis",
        ylabel="y-axis");
ax2.set(title="Scatter Plot",
        xlabel="x-axis",
        ylabel="y-axis");
ax3.set(title="Bar Plot",
        xlabel="x-axis",
        ylabel="y-axis");
ax4.set(title="Histogram",
        xlabel="x-axis",
        ylabel="y-axis");
```

# Great use above plotting mechanism when you explore real data

```python
# OO method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
             x="age",
             y="chol",
             c="target",
             ax=ax);
# ax.set_xlim([45, 100]);
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
scatter = ax.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);
# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="Cholesterol");
# Add a legend
ax.legend(*scatter.legend_elements(), title="Target");
ax.axhline(over_50["chol"].mean(),
           linestyle="--");
```

```python
# let's create subplots of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10),
                               sharex=True)
# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])
# Customize ax0
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="Cholesterol")
# Add a legend to ax0
ax0.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
            linestyle="--");
# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])
# Customize ax1
ax1.set(title="Heart Disease and Max Heart Rate",
        xlabel="Age",
        ylabel="Max Heart Rate")
# Add a legend to ax1
ax1.legend(*scatter.legend_elements(), title="Target")
# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
            linestyle="--");
# Add a title to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
